```python
import pandas as pd
import matplotlib.pyplot as plt
import random
from sklearn.metrics import accuracy_score
```

```python
def mode_ties(s):
    m = s.mode()
    if len(m) > 1:
        random.shuffle(m)
        return(m[0])
    return m
```

```python
tasks = ['adult_ordinal', 'waterbird', 'dog', 'face', 'movie', 'temporal', 'weather_ordinal']
for task in tasks:
    df_ds = pd.read_csv('results_supervised_' + task + '.csv')[['level', 'accuracy']]
    df_ds.columns = ['supervision_level', 'ds']
    df_ds = df_ds.groupby('supervision_level').mean()
    df_ds.index = df_ds.index.astype(str)
    ax = df_ds.plot.line(title=task)
    ax.set_ylabel('accuracy')
    ax.set_ylim(.5,1)
```
